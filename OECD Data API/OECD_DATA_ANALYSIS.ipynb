```python
# Filter for the year 2015
df_2015 = df[df['TimeDim'] == 2015.0]

# Convert 'PercentHouseholdExp' to numeric values
df_2015['PercentHouseholdExp'] = pd.to_numeric(df_2015['PercentHouseholdExp'], errors='coerce')

# Drop rows with NaN values in required columns
df_2015 = df_2015.dropna(subset=['2010 index Value Alc', '2010 index Value All', 'PercentHouseholdExp'])

# Debugging
print(df_2015)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_2015['2010 index Value Alc'], df_2015['2010 index Value All'], s=df_2015['PercentHouseholdExp'] * 10, alpha=0.5)

# Adding labels for each bubble
for i, txt in enumerate(df_2015['SpatialDim']):
    plt.text(df_2015['2010 index Value Alc'].iloc[i], df_2015['2010 index Value All'].iloc[i], txt, ha='center', va='center')

# Adding labels and title
plt.xlabel('2010 Index Value Alc')
plt.ylabel('2010 Index Value All')
plt.title('Bubble Graph of 2010 Index Values vs Total Alcohol Market Size for 2015')

# Show plot
plt.show()
```

```python
# Filter for the year 2015
df_2015 = df[df['TimeDim'] == 2015]

# Convert 'PercentHouseholdExp' to numeric values and multiply by 100
df_2015['PercentHouseholdExp'] = pd.to_numeric(df_2015['PercentHouseholdExp'], errors='coerce') * 500

# Drop rows with NaN values in required columns
df_2015 = df_2015.dropna(subset=['2010 index Value Alc', '2010 index Value All', 'PercentHouseholdExp'])

# Plotting
plt.figure(figsize=(10, 6))
colors = ['darkred' if country in ['AUS', 'CAN', 'EST'] else 'blue' for country in df_2015['SpatialDim']]
plt.scatter(df_2015['2010 index Value Alc'], df_2015['2010 index Value All'], s=df_2015['PercentHouseholdExp'] * 10, alpha=0.5, c=colors)

# Adding labels for each bubble
for i, txt in enumerate(df_2015['SpatialDim']):
    plt.text(df_2015['2010 index Value Alc'].iloc[i], df_2015['2010 index Value All'].iloc[i], txt, ha='center', va='center')

# Adding labels and title
plt.xlabel('Household Spending Growth of Alcohol', color='white')
plt.ylabel('Household Spending Growth of All Commodities', color='white')
plt.title('Household Spending Growth 2010-2015, % of Household Spend on Alcohol', color='white')

# Set background to transparent
plt.gcf().set_facecolor('none')
plt.gca().set_facecolor('none')

# Set all lines, ticks, and labels to white
plt.rcParams['axes.edgecolor'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'

# Save the plot as a PNG file
plt.savefig('household_spend_scatter.png', transparent=True)

# Show plot
plt.show()
```

```python
# Filter for the year 2015
df_2015 = df[df['TimeDim'] == 2015.0]

# Convert 'Total Household Exp' to numeric values, coerce errors to NaN
df_2015['Total Household Exp'] = pd.to_numeric(df_2015['Total Household Exp'], errors='coerce')

# Drop rows with NaN values in required columns after conversion
df_2015 = df_2015.dropna(subset=['2010 index Value Alc', '2010 index Value All', 'PercentHouseholdExp', 'Total Household Exp'])

# Calculate the bubble size
df_2015['BubbleSize'] = (df_2015['PercentHouseholdExp'] * df_2015['Total Household Exp']) / 60000000000

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df_2015['2010 index Value Alc'], df_2015['2010 index Value All'], s=df_2015['BubbleSize'], alpha=0.5)

# Adding labels for each bubble
for i, txt in enumerate(df_2015['SpatialDim']):
    plt.text(df_2015['2010 index Value Alc'].iloc[i], df_2015['2010 index Value All'].iloc[i], txt, ha='center', va='center')

# Adding labels and title
plt.xlabel('2010 Index Value Alc')
plt.ylabel('2010 Index Value All')
plt.title('Bubble Graph of 2010 Index Values vs Total Alcohol Market Size for 2015')

# Show plot
plt.show()
```
