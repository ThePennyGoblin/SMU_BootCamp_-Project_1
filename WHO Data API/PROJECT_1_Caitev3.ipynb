```python
#define data to pull from WHO GHO API
url= "https://ghoapi.azureedge.net/api/SA_0000001476"
url1= "https://ghoapi.azureedge.net/api/SA_0000001398"
url2= "https://ghoapi.azureedge.net/api/SA_0000001732"
url3= "https://ghoapi.azureedge.net/api/SA_0000001523"
url4= "https://ghoapi.azureedge.net/api/WHOSIS_000002"
url5= "https://ghoapi.azureedge.net/api/SA_0000001822"
url6= "https://ghoapi.azureedge.net/api/SA_0000001398"
url7= "https://ghoapi.azureedge.net/api/TAXBEV_MOSTSOLDSPIRIT"
url8= "https://ghoapi.azureedge.net/api/SA_0000001400_ARCHIVED"
url9= "https://ghoapi.azureedge.net/api/SA_0000001688"
url10= "https://ghoapi.azureedge.net/api/SA_0000001398_ARCHIVED"
```

```python
# list to store URLS 
urls= [url,url1,url2,url3,url4,url5,url6,url7,url8, url9, url10]

# Dictionary to store dataframes
dataframes = {}

# Fetch data from each URL and create dataframes
for idx, url in enumerate(urls):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        df_name = f"df_{idx}"  # Name the dataframe
        dataframes[df_name] = pd.DataFrame(data)
        print("Data from", url, "has been fetched and stored in", df_name)
    else:
        print("Failed to fetch data from", url)
```

```python
# Assuming dataframes is a dictionary containing DataFrames
for df_name, df in dataframes.items():
    # Expand dictionaries into separate columns and rows
    expanded_df = pd.json_normalize(df['value'])  # Replace 'your_column_name_here' with the actual column name
    
    # Assign the expanded DataFrame back to the original variable
    globals()[df_name] = expanded_df
    
    # Write DataFrame to CSV
    csv_filename = f"{df_name}.csv"
    expanded_df.to_csv(csv_filename, index=False)  # Fixing the syntax here
    print(f"DataFrame '{df_name}' has been written to '{csv_filename}' CSV file.")
    print()
```

```python
df_1clean= df_1[['SpatialDim', 
                 'ParentLocationCode', 
                 'TimeDim','Dim1', 
                 'Value', 
                 'TimeDimensionValue']]
df_1clean.columns = ['SpatialDim', 
                     'ParentLocationCode', 
                     'TimeDim',
                     'BevType', 
                     'PercBevType', 
                     'TimeDimensionValue']
display(df_1clean)
```

```python
# Set cells with "<" to null in the 'LPerCapita' column
df_1clean['PercBevType'] = df_1clean['PercBevType'].replace('<1', 0)
df_1clean['PercBevType'] = df_1clean['PercBevType'].replace('No data', 0)
```

```python
# Create new columns for SPIRITS, BEER, WINE, and OTHER_ALCOHOL percentages
df_1clean['SPIRITS_PERCENT'] = 0
df_1clean['BEER_PERCENT'] = 0
df_1clean['WINE_PERCENT'] = 0
df_1clean['OTHER_ALCOHOL_PERCENT'] = 0

# Iterate over rows and update SPIRITS, BEER, WINE, or OTHER_ALCOHOL columns based on 'BevType'
for index, row in df_1clean.iterrows():
    if row['PercBevType'] is not None and row['PercBevType'] != 'No data':
        if 'SPIRITS' in row['BevType']:
            df_1clean.at[index, 'SPIRITS_PERCENT'] = float(row['PercBevType'])
        elif 'BEER' in row['BevType']:
            df_1clean.at[index, 'BEER_PERCENT'] = float(row['PercBevType'])
        elif 'WINE' in row['BevType']:
            df_1clean.at[index, 'WINE_PERCENT'] = float(row['PercBevType'])
        elif 'OTHER_ALCOHOL' in row['BevType']:
            df_1clean.at[index, 'OTHER_ALCOHOL_PERCENT'] = float(row['PercBevType'])

# Drop the original 'PercBevType' and 'BevType' columns if no longer needed
df_1clean.drop(columns=['PercBevType', 'BevType'], inplace=True)

# Display the updated DataFrame
display(df_1clean)
```

```python
# Group by 'SpatialDim' and aggregate the values of 'SPIRITS', 'BEER', 'WINE', and 'OTHER_ALCOHOL'
df_1clean1 = df_1clean.groupby(['SpatialDim', 
                                'ParentLocationCode', 
                                'TimeDim', 
                                'TimeDimensionValue']).agg({'SPIRITS_PERCENT': 'sum', 
                                                            'BEER_PERCENT': 'sum', 
                                                            'WINE_PERCENT': 'sum', 
                                                            'OTHER_ALCOHOL_PERCENT': 'sum'}).reset_index()

# Display the updated DataFrame
display(df_1clean1)
```
